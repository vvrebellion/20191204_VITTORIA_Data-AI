```

#import random
#classes = ['angry', 'happy', 'sad','neutral']
#predictedClass = random.choice(classes)
#predictedClass = 'neutral'
if predictedClass == 'angry':
    valence = '150000'
    arousal = '750000'
elif predictedClass == 'happy': #update after testing
    valence = '850000'
    arousal = '850000'
elif predictedClass == 'sad':
    valence = '140000'
    arousal = '120000'
#elif predictedClass == 'neutral':
#    valence = '425000'
#    arousal = '425000'

url = 'http://musicovery.com/api/V6/playlist.php?&fct=getfrommood&popularitymax=100&popularitymin=50&starttrackid=&date70=true&trackvalence='+valence+'&trackarousal='+arousal+'&resultsnumber=15'
data = requests.get(url).json()
#print(data)
 
tracks = data['tracks']
track = tracks['track']
#track

title = (list(m['title'] for m in track))
artist = (list(m['artist_display_name'] for m in track))

title_list = pd.DataFrame(title, columns = ['title'])
title_list['artist'] = artist

print('If you are feeling ' + predictedClass + '...')
print('How about listening one of these songs: ')
title_list
```

```
a = input('Choose the number of the song you want to linsten to: ')
a=int(a)

songname = title[a] + ' ' + artist[a]
songname

from youtube_search import YoutubeSearch
results = YoutubeSearch(songname, max_results=10).to_dict()
#print(results)

title = (list(m['title'] for m in results))
url = (list(m['url_suffix'] for m in results))

title_list = pd.DataFrame(url, columns = ['url'])
title_list['title'] = title

print('We found these results on youtube:')
title_list
```
